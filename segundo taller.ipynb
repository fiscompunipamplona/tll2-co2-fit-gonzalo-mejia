```python
dataco2= loadtxt("co2_mm_gl.txt",float)
X=array(dataco2[:,2])
Y=array(dataco2[:,3])
X1=array(dataco2[0:235,2])
Y1=array(dataco2[0:235,3])
X2=array(dataco2[235:470,2])
Y2=array(dataco2[235:470,3])
plot(X,Y,alpha=1,c="green")
#plot(X1,Y1,alpha=1)
X=array([np.ones(470),X]).T
X1=array([np.ones(235),X1]).T
X2=array([np.ones(235),X2]).T
B=np.linalg.inv(X.T @ X) @ X.T @ Y
B1=np.linalg.inv(X1.T @ X1) @ X1.T @ Y1
B2=np.linalg.inv(X2.T @ X2) @ X2.T @ Y2
plot([1980,2019],[B[0]+B[1]*1980,B[0]+B[1]*2019],c="red")
plot([1980,2000],[B1[0]+B1[1]*1980,B1[0]+B1[1]*2000],c="blue")
plot([2000,2019],[B2[0]+B2[1]*2000,B2[0]+B2[1]*2019],c="black")
show()
```

Podemos observamos que el CO2 cada año decimal va aumentado su promedio mensual global, es decir, cada año la atmosfera se ve más contaminada, sin importar la lucha contra la contaminación.
